```python
# If the datafile does not have column headings

file_path = file_dir + "/friends1.csv"

customschema=StructType([
    StructField("seq",IntegerType(), True),
    StructField("Name",StringType(), True),
    StructField("age",IntegerType(), True),
    StructField("friend_count",IntegerType(), True)    
])

nohead_df = spark.read.csv(file_path, header=False, sep=",", schema=customschema)
```

```python
nohead_df.printSchema()
nohead_df.show()
```

```python
# Write a dataframe into a Parquet file

# csv_df.write.parquet("friends1.parquet")

# If the file already exists, overwrite it

csv_df.write.mode("overwrite").parquet("friends1.parquet")
```

```python
# Read a Parquet file into a DataFrame

parquet_df = spark.read.parquet("friends1.parquet")
```

```python
parquet_df.printSchema()
parquet_df.show()
```

```python
file_path2 = file_dir + "/income.csv"

csv2_df = spark.read.option("header", "true").option("delimiter",",").csv(file_path2)
```

```python
csv2_df.printSchema()
csv2_df.show(5)
```

```python
# Convert a Parquet DataFrame into a SQL View and use SQL instead

# Parquet files can also be used to create a temporary view and then used in SQL statements.
csv2_df.createOrReplaceTempView("temp_view")
alabama = spark.sql("SELECT id,city,State_Name FROM temp_view WHERE State_ab = 'AL'")
```

```python
alabama.show(5)
```

```python
#Use partitionBy() to split the date into partitions (by state), and write it

csv2_df.write.option("header",True) \
        .partitionBy("State_ab") \
        .mode("overwrite") \
        .csv(file_dir + "/by_state")
```
